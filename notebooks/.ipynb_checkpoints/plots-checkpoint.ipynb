```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from collections import *
%matplotlib inline  
```

```python
x_data = np.arange(10)
width = 0.25
x_data - width
```

```python
y1_data = np.linspace(1,14,10)
y2_data = np.linspace(1,13,10)
y3_data = np.linspace(1,12,10)
```

```python
# plt.xkcd(randomness=105,scale=5,length=550)
plt.tight_layout()
plt.rcdefaults()
```

```python
fig = plt.figure()
ax = fig.add_subplot()
```

```python
plt.barh(x_data-width,y1_data,height=width,color="blue",label="Blue")
plt.barh(x_data,y2_data,height=width,color="red",label="Red")
plt.barh(x_data+width,y3_data,height=width,color="green",label="Green")
plt.legend()
```

```python
plt.pie([3,1],labels=["Pacman","Background"],
        wedgeprops={"edgecolor": "black"},
       colors=["yellow","black","blue","green"],
       shadow=True,startangle=45)
plt.title("Pacman Illustration")
```

```python
plt.style.use("fivethirtyeight")

minutes = [1,2,3,4,5,6,7,8,9]
player1 = [1,1,2,2,2,3,4,4,4]
player2 = [1,1,1,2,2,2,2,2,3]
player3 = [1,2,2,2,3,3,4,5,5]

fig, ax = plt.subplots()

labels = ["player1","player2","player3"]

ax.stackplot(minutes,player1,player2,player3,labels=labels)
plt.legend(loc=2)
plt.title("Players' Scores Over 9 Minutes")

plt.tight_layout()
```

```python
plt.style.use("fivethirtyeight")

minutes = [1,2,3,4,5,6,7,8,9]

labels = ["employee1","employee2","employee3"]
player1 = [8,6,5,5,4,3,2,1,0]
player2 = [0,1,1,2,2,3,3,4,4]
player3 = [0,1,2,1,2,2,3,3,4]
fig, ax = plt.subplots()

ax.stackplot(minutes,player1,player2,player3,labels=labels)
ax.legend(loc=(0.07,0.09))
plt.title("Employees' Work Load Over 9 Days")

plt.tight_layout()
```

```python
x_data = [2,1,5,4,3,6,1,3,4,3,1,4,2,1,4,5,3,6]
y_data = [4,3,2,1,2,4,6,4,3,5,6,1,2,6,6,4,3,2]
colors = list(map(lambda a,b: a+b-4,x_data,y_data))
plt.scatter(x_data,y_data,c=colors,
           cmap="Greens",edgecolor="black",
           linewidth=1,alpha=0.75)
cbar = plt.colorbar()
cbar.set_label("Satisfaction")
```
