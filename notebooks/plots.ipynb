```python
plt.style.use("default")
plt.xkcd(randomness=1)
browsers = ["Chrome","Safari","Firefox","Others"]
usageRates = [63.54,19.24,3.79,100-63.54-19.24-3.79]
# explode firefox 
explode = (0,0,0.2,0,)
plt.pie(usageRates,labels=browsers,
        wedgeprops={"edgecolor": "black"},
       shadow=True,startangle=45,
       explode=explode,autopct='%1.f%%')
plt.title("2021 Browser Popularity")
```

```python
plt.style.use("fivethirtyeight")

minutes = [1,2,3,4,5,6,7,8,9]
player1 = [1,1,2,2,2,3,4,4,4]
player2 = [1,1,1,2,2,2,2,2,3]
player3 = [1,2,2,2,3,3,4,5,5]

fig, ax = plt.subplots()

labels = ["player1","player2","player3"]

ax.stackplot(minutes,player1,player2,player3,labels=labels)
plt.legend(loc=2)
plt.title("Players' Scores Over 9 Minutes")

plt.tight_layout()
```

```python
plt.rcdefaults()
plt.style.use("fivethirtyeight")

minutes = [1,2,3,4,5,6,7,8,9]

labels = ["employee1","employee2","employee3"]
player1 = [8,6,5,5,4,3,2,1,0]
player2 = [0,1,1,2,2,3,3,4,4]
player3 = [0,1,2,1,2,2,3,3,4]
colors=["Green","Orange","Grey"]
fig, ax = plt.subplots()

ax.stackplot(minutes,player1,player2,player3,labels=labels,colors=colors)
ax.legend(loc=(0.07,0.09))
plt.title("Employees' Work Load Over 9 Days")

plt.tight_layout()
```

```python
plt.rcdefaults()
plt.style.use("seaborn")
x_data = [2,1,5,4,3,6,1,3,4,3,1,4,2,1,4,5,3,6]
y_data = [4,3,2,1,2,4,6,4,3,5,6,1,2,6,6,4,3,2]
colors = list(map(lambda a,b: a+b-4,x_data,y_data))
plt.scatter(x_data,y_data,c=colors,
           cmap="viridis_r", edgecolor="black",
           linewidth=1,alpha=0.75)
cbar = plt.colorbar()
cbar.set_label("Satisfaction")
```

```python
plt.rcdefaults()
plt.style.use("fivethirtyeight")
ages = [25,28,39,19,48,19,48,68,23,39,48,9,29,48,51,55,38,29]
bins = [10,20,30,40,50,60,70]

plt.hist(ages,bins=bins,edgecolor="black",color="green")
median_age = sum(ages) / len(ages)
color = "red"

plt.axvline(median_age,color=color,label="median age")
plt.legend()
```

```python
plt.xkcd()
plt.style.use("fivethirtyeight")
years = [1960,1970,1980,1990,2000,2010,2020]
produced = [28_000,100_000,5_360_000,14_000_000,168_000_000,7_500_000_000,40_000_000_000]

fig, ax = plt.subplots()
ax.plot(years,produced,color="black",label="PC Production")
ax.set_title("Personal Computer Production Over The Last 70 Years")
ax.set_xlabel("Years")
ax.set_ylabel("Computers Produced")
ax.fill_between(years,produced,min(produced),color="green",alpha=0.3)
ax.fill_between(years,produced,max(produced),color="red",alpha=0.3)

plt.legend()
plt.yscale(value="log")
plt.figtext(0.5,-0.15,"The values are made up", ha="center", fontsize=18, bbox={"facecolor":"orange", "alpha":0.5, "pad":5})
```
